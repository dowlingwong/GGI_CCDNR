```python
from scipy.signal import find_peaks
import numpy as np
import matplotlib.pyplot as plt

# Histograma en el rango bajo (ya lo hiciste, lo repito por claridad)
counts, bin_edges = np.histogram(all_pixels, bins=400, range=(0, 5000))
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

# Trabajamos solo con esta regin
x = bin_centers
y = counts

plt.figure(figsize=(8,6))
plt.plot(x, y)
plt.yscale('log')
plt.xlabel("Pixel value")
plt.ylabel("Nmero de pxeles")
plt.title("Regin baja con estructura de picos")
plt.show()

# Buscar picos: ajusta 'prominence' si hace falta (ms grande = menos picos falsos)
peaks_idx, props = find_peaks(y, prominence=50)

peak_positions = x[peaks_idx]

print("Posiciones de picos encontrados (primeros 10):")
print(peak_positions[:10])
```
